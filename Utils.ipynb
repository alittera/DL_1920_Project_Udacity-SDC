```python
import cv2
import math
import matplotlib.pyplot as plt
import glob
from PIL import Image 
from livelossplot import PlotLosses
```

```python
def draw_angles_image(path, save_path, a,b, save=True, show=False):

    o_a = a
    o_b = b
    img = cv2.imread(path)
    height, width = img.shape[:2]
    
    thick = int(width/70)
    
    #print(height, width)
    
    off_x = int(width/2) 
    off_y = int(height)

    L_a = int(height/3)
    L_b = int(L_a/1.8)
    
    

    a = (a+270)*(math.pi / 180)
    b = (b+270)*(math.pi / 180)


    x_a = int((L_a* math.cos(a))+ off_x)
    y_a = int((L_a* math.sin(a))+ off_y)

    x_b = int((L_b* math.cos(b))+ off_x)
    y_b = int((L_b* math.sin(b))+ off_y)

    img = cv2.line(img,(off_x,off_y),(x_a,y_a),(255,0,0),thick)
    img = cv2.line(img,(off_x,off_y),(x_b,y_b),(0,0,255),thick)
    
    font                     = cv2.FONT_HERSHEY_SIMPLEX
    bottomLeftCornerOfText_1 = (width-int(width/6), int(height/11) )
    bottomLeftCornerOfText_2 = (width-int(width/6), int(height/6) )
    bottomLeftCornerOfText_3 = (int(width/14), int(height/11) )
    bottomLeftCornerOfText_4 = (int(width/14), int(height/6) )
    fontScale                = 0.8
    fontColor_1              = (255,0,0)
    fontColor_2              = (0,0,255)
    lineType                 = 2

    cv2.putText(img, "{0:.2f}".format(o_a), 
    bottomLeftCornerOfText_1, 
    font, 
    fontScale,
    fontColor_1,
    lineType)
    
    cv2.putText(img, "{0:.2f}".format(o_b), 
    bottomLeftCornerOfText_2, 
    font, 
    fontScale,
    fontColor_2,
    lineType)
    
    cv2.putText(img, "True", 
    bottomLeftCornerOfText_3, 
    font, 
    fontScale,
    fontColor_1,
    lineType)
    
    cv2.putText(img, "Pred", 
    bottomLeftCornerOfText_4, 
    font, 
    fontScale,
    fontColor_2,
    lineType)
    
 
    
    if save:
        cv2.imwrite(save_path, img)
        
    if show:
        plt.imshow(img)
        plt.show()
```

```python
import gc
def video_from_frames(folder, out_filename, extension="jpg", ):

    img_array = []
    files = glob.glob(folder + '/*.' + extension)
    files.sort()
    for filename in files:
        img = cv2.imread(filename)
        height, width, layers = img.shape
        size = (width,height)
        img_array.append(img)


    out = cv2.VideoWriter(out_filename,cv2.VideoWriter_fourcc(*'DIVX'), 22, size)

    for i in range(len(img_array)):
        out.write(img_array[i])
    out.release()
    
    del img_array
    del out
    
    gc.collect()
```
